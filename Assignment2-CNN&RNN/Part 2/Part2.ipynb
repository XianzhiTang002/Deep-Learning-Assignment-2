```python
import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt

from torchvision import datasets, transforms
from torch.utils.data import DataLoader

from cnn_model import CNN
from cnn_train import train
```

```python
LEARNING_RATE_DEFAULT = 1e-4
BATCH_SIZE_DEFAULT = 32
MAX_EPOCHS_DEFAULT = 20
EVAL_FREQ_DEFAULT = 2
```

```python
train_loss, train_acc, test_loss, test_acc = train(
    learning_rate=LEARNING_RATE_DEFAULT,
    max_steps=MAX_EPOCHS_DEFAULT,
    eval_freq=EVAL_FREQ_DEFAULT,
    batch_size=BATCH_SIZE_DEFAULT)
```

```python
plt.figure()
plt.plot(range(0, MAX_EPOCHS_DEFAULT + 1, EVAL_FREQ_DEFAULT), train_loss, label='Train Loss')
plt.plot(range(0, MAX_EPOCHS_DEFAULT + 1, EVAL_FREQ_DEFAULT), test_loss, label='Test Loss')
plt.title('CNN Train and Test Loss over Epochs on CIFAR10')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.grid()
plt.show()
```

```python
plt.figure()
plt.plot(range(0, MAX_EPOCHS_DEFAULT + 1, EVAL_FREQ_DEFAULT), train_acc, label='Train Accuracy')
plt.plot(range(0, MAX_EPOCHS_DEFAULT + 1, EVAL_FREQ_DEFAULT), test_acc, label='Test Accuracy')
plt.title('CNN Train and Test Accuracy over Epochs on CIFAR10')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.ylim(0, 1)
plt.legend()
plt.grid()
plt.show()
```
